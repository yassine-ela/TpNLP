```python
df=pd.read_csv('/content/movie_review.csv')
df
```

```python
import string
# Tlcharger la liste des stop words (si ce n'est pas dj fait)
nltk.download('stopwords')

# Fonction pour effectuer le prtraitement du texte
def preprocess_text(text):
    # Supprimer la ponctuation
    text = ''.join([char for char in text if char not in string.punctuation])
    # Convertir en minuscules
    text = text.lower()
    # Supprimer les stop words
    stop_words = set(stopwords.words('english'))
    tokens = text.split()
    text = ' '.join([word for word in tokens if word not in stop_words])
    return text

# Appliquer la fonction de prtraitement  la colonne "Text" de votre DataFrame
df['text'] = df['text'].apply(preprocess_text)
#vecteur par sentence
df
# Maintenant, la colonne "Text" de votre DataFrame contient les donnes textuelles prtraites
```

```python
from gensim.models import Word2Vec

# Votre DataFrame 'df' contenant les colonnes 'text' et 'tag'
# Assurez-vous que le prtraitement a dj t appliqu  la colonne 'text'

# Obtenez une liste de listes de mots  partir de la colonne 'text'
preprocessed_data = df['text'].apply(lambda x: x.split())

# Entraner le modle Word2Vec
model = Word2Vec(sentences=preprocessed_data, vector_size=100, window=5, min_count=1, sg=0)

# Arguments :
# - sentences : Les donnes textuelles prtraites.
# - vector_size : La dimension des vecteurs de mots (taille des embeddings).
# - window : La taille de la fentre de contexte pour l'apprentissage.
# - min_count : Le nombre minimum d'occurrences d'un mot pour tre pris en compte.
# - sg : Utiliser Skip-gram (1) ou CBOW (0).

# Sauvegarder le modle
model.save("votre_modele_word2vec.model")
```

```python
import numpy as np
from gensim.models import Word2Vec

# Charger le modle Word2Vec entran
model = Word2Vec.load("votre_modele_word2vec.model")

# Votre DataFrame 'df' contenant les colonnes 'text' et 'tag'
# Assurez-vous que le prtraitement a dj t appliqu  la colonne 'text'

# Fonction pour obtenir le vecteur moyen d'une revue
def get_review_vector(review, model):
    words = review.split()
    word_vectors = []
    for word in words:
        if word in model.wv:
            word_vectors.append(model.wv[word])
    if len(word_vectors) > 0:
        return np.mean(word_vectors, axis=0)
    else:
        return np.zeros(model.vector_size)  # Vecteur nul si aucun mot trouv

# Appliquer la fonction pour obtenir les vecteurs de toutes les revues
df['review_vector'] = df['text'].apply(lambda x: get_review_vector(x, model))

# Maintenant, chaque revue est reprsente par un vecteur dans la colonne 'review_vector'
```

```python
from sklearn.model_selection import train_test_split

# Votre DataFrame 'df' contenant les donnes
# 'X' contient les vecteurs des revues, 'y' contient les tiquettes ('pos' ou 'neg')

# Diviser les donnes en ensembles d'entranement et de test
X_train, X_test, y_train, y_test = train_test_split(df['review_vector'].tolist(), df['tag'], test_size=0.2, random_state=42)

# 'X_train' et 'X_test' contiennent les vecteurs des revues pour les ensembles d'entranement et de test
# 'y_train' et 'y_test' contiennent les tiquettes correspondantes

# Vous pouvez ajuster le paramtre 'test_size' pour dfinir la proportion de donnes  utiliser pour les tests
# 'random_state' permet de garantir la reproductibilit de la division (utilisez une valeur fixe)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC

# Utilisez le modle de rgression logistique (Logistic Regression) comme exemple.
# Vous pouvez utiliser le modle de votre choix en remplaant LogisticRegression par RandomForestClassifier ou SVC.

# Crer un modle
model = LogisticRegression()

# Entraner le modle sur l'ensemble d'entranement
model.fit(X_train, y_train)

# Prdictions sur l'ensemble de test
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report, confusion_matrix

# Calculer l'accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy : {accuracy:.2f}")

# Calculer la prcision
precision = precision_score(y_test, y_pred, average='weighted')  # weighted pour la moyenne pondre
print(f"Prcision : {precision:.2f}")

# Calculer le recall
recall = recall_score(y_test, y_pred, average='weighted')  # weighted pour la moyenne pondre
print(f"Recall : {recall:.2f}")

# Calculer le F1-score
f1 = f1_score(y_test, y_pred, average='weighted')  # weighted pour la moyenne pondre
print(f"F1-Score : {f1:.2f}")

# Afficher un rapport de classification dtaill
classification_rep = classification_report(y_test, y_pred)
print("Rapport de classification :\n", classification_rep)

# Afficher la matrice de confusion
confusion = confusion_matrix(y_test, y_pred)
print("Matrice de confusion :\n", confusion)
```
